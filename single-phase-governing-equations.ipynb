## GENERAL GOVERNING EQUATIONS
***

### Continuity Equation
    
* Mass conservation requires:
    
    \begin{equation}\text{Accumulation } + [\text{Out} - \text{In}] = 0\end{equation}

* Accumulation:
    
    \begin{equation}\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]\Delta x \Delta y \Delta z}{\Delta t}\end{equation}
    
* Net outflow through x:
    
    \begin{equation}\left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]\Delta y \Delta z\end{equation}
    
* Net outflow through y:

    \begin{equation}\left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]\Delta x \Delta z\end{equation}
    
* Net outflow through z:

    \begin{equation}\left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]\Delta x \Delta y\end{equation}
    
* Combining all terms:
    
    \begin{equation}\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]\Delta x \Delta y \Delta z}{\Delta t} + \left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]\Delta y \Delta z + \left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]\Delta x \Delta z + \left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]\Delta x \Delta y = 0\end{equation}

* Dividing through by the volume of the fluid element, $\Delta x \Delta y \Delta z$:

    \begin{equation}\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]}{\Delta t} + \frac{\left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]}{\Delta x} + \frac{\left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]}{\Delta y} + \frac{\left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]}{\Delta z} = 0\end{equation}
    
* In the limits of $\Delta t$, $\Delta x$, $\Delta y$, and $\Delta z$ approaching zero:

    \begin{equation}\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z} = 0\end{equation}
    
### Momentum Balance Equations


* The rate of change of a conserved quantity $\phi\left(t, x, y, z\right)$ following a fluid particle is given by the total/material/substantial derivative of the quantity.

    \begin{equation}\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + \frac{\partial \phi}{\partial x} \frac{\partial x}{\partial t} + \frac{\partial \phi}{\partial y}\frac{\partial y}{\partial t} + \frac{\partial \phi}{\partial z}\frac{\partial z}{\partial t}\end{equation}
    
* Recognizing that $\frac{\partial x}{\partial t} = u_x$, $\frac{\partial y}{\partial t} = u_y$, and $\frac{\partial z}{\partial t} = u_z$:
    
    \begin{equation}\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y \frac{\partial \phi}{\partial y} + u_z \frac{\partial \phi}{\partial z}\end{equation}

* The above equation describes the rate of change per unit mass using Lagrangian approach - the rate of change per unit volume (Eulerian approach) can be obtained by multiplying the prior equation by the density:
    
    \begin{equation}\rho \frac{D\phi}{Dt} = \rho \left[ \frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y \frac{\partial \phi}{\partial y} + u_z \frac{\partial \phi}{\partial z} \right]\end{equation}
    
* It can alternatively be derived by considering a fluid element fixed in space. The total rate of change for the fluid element comprises of time rate of change and advection from the surroundings through the boundaries of the fluid element.
    
    \begin{equation}\text{Total Rate of Change} = \text{Time Rate of Change} + \text{Advection from Surroundings}\end{equation}
    
    \begin{equation}\rho \Delta x \Delta y \Delta z \frac{D\phi}{Dt} = \frac{\left[\rho \phi|_{t + \Delta t} - \rho \phi|_t\right]\Delta x \Delta y \Delta z}{\Delta t} + \left[\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x\right]\Delta y \Delta z + \left[\rho u_y \phi|_{y + \Delta y} - \rho u_y \phi|_y\right] \Delta x \Delta z + \left[\rho u_z \phi|_{z + \Delta z} - \rho u_z \phi|_z\right] \Delta x \Delta y\end{equation}
    
     \begin{equation}\rho \frac{D\phi}{Dt} = \frac{\left[\rho \phi|_{t + \Delta t} - \rho \phi|_t\right]}{\Delta t} + \frac{\left[\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x\right]}{\Delta x} + \frac{\left[\rho u_y \phi|_{y + \Delta y} - \rho u_y \phi|_y\right]}{\Delta y} + \frac{\left[\rho u_z \phi|_{z + \Delta z} - \rho u_z \phi|_z\right]}{\Delta z}\end{equation}
    
    \begin{equation}\rho \frac{D\phi}{Dt} = \frac{\partial \left(\rho \phi\right)}{\partial t} + \frac{\partial \left(\rho u_x \phi\right)}{\partial x} + \frac{\partial \left(\rho u_y \phi\right)}{\partial y} + \frac{\partial \left(\rho u_z \phi\right)}{\partial z}\end{equation}
    
     \begin{equation}\rho \frac{D\phi}{Dt} = \rho \left[\frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}\right] + \phi\left[\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z}\right]\end{equation}
     
    \begin{equation}\rho \frac{D\phi}{Dt} = \rho \left[\frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}\right]\end{equation}
     
* From Newton's second law of motion, the rate of change of momentum (conserved quantity) is equal to the sum of forces acting on the fluid element:
    
    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = \sum F_x\end{equation}
    
    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = \sum F_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = \sum F_z\end{equation}

* Forces (per unit volume) under consideration:
    
    1. Pressure forces: 

        * x-component: $-\frac{\partial p}{\partial x}$

        * y-component: $-\frac{\partial p}{\partial y}$

        * z-component: $-\frac{\partial p}{\partial z}$
    
    2. Viscous shear forces (first subscript is the cordinate perpendicular to the force and second subscript is the direction in which the force is acting): 

        * x-components: $\frac{\partial \tau_{xx}}{\partial x}$, $\frac{\partial \tau_{yx}}{\partial x} $, $\frac{\partial \tau_{zx}}{\partial x}$
        
        * y-components:   $\frac{\partial \tau_{xy}}{\partial y}$, $\frac{\partial \tau_{yy}}{\partial y}$, $\frac{\partial \tau_{zy}}{\partial y}$
        
        * z-components:  $\frac{\partial \tau_{xz}}{\partial z}$, $\frac{\partial \tau_{yz}}{\partial z} $, $\frac{\partial \tau_{zz}}{\partial z}$

    3. Gravity: 
    
        * x-component: $\rho g_x$
        
        * y-component: $\rho g_y$ 
        
        * z-component: $\rho g_z$

    4. Other forces: 
        
        * x-component: $S_x$ 
        
        * y-component: $S_y$ 
        
        * z-component: $S_z$

* Plugging in the forces:

    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial \tau_{xx}}{\partial x} + \frac{\partial \tau_{yx}}{\partial x} + \frac{\partial \tau_{zx}}{\partial x} + \rho g_x + S_x\end{equation}
    
    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial \tau_{xy}}{\partial y} + \frac{\partial \tau_{yy}}{\partial y} + \frac{\partial \tau_{zy}}{\partial y} + \rho g_y + S_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial \tau_{xz}}{\partial z} + \frac{\partial \tau_{yz}}{\partial z} + \frac{\partial \tau_{zz}}{\partial z} + \rho g_z + S_z\end{equation}

### Energy Equation

* The equation for conservation of energy is governed by the first law of thermodynamics, which in current context states that the rate of change of energy for a fluid element is equal to the sum of heat added to the fluid element and the work done on the fluid element:
    
    \begin{equation} \text{Rate of Change of Energy for a Fluid Element} = \text{Rate of Heat Added to Fluid Element} + \text{Rate of Work Done on Fluid Element}\end{equation}
    
* The rate of change of energy for the fluid element is given by:
    
    \begin{equation} \rho \Delta x \Delta y \Delta z \frac{DE}{Dt}\end{equation}

* The rate of heat added to the fluid element is given by:
    
    \begin{equation} \text{X:} -\frac{\partial q_x}{\partial x} \Delta x \Delta y \Delta z = -\frac{\partial}{\partial x} \left(k \frac{\partial T}{\partial x}\right) \Delta x \Delta y \Delta z \end{equation}
    
    \begin{equation} \text{Y:} -\frac{\partial q_y}{\partial y} \Delta x \Delta y \Delta z = -\frac{\partial}{\partial y} \left(k \frac{\partial T}{\partial y}\right) \Delta x \Delta y \Delta z\end{equation}
    
    \begin{equation} \text{Z:} -\frac{\partial q_z}{\partial z} \Delta x \Delta y \Delta z = -\frac{\partial}{\partial z} \left(k \frac{\partial T}{\partial z}\right) \Delta x \Delta y \Delta z\end{equation}
    
* The rate of work done on the fluid element by the shear forces is given by the product of the shear force and associated velocity:
    
    \begin{equation} \text{X: } \left[\frac{\partial \left(u_x \tau_{xx}\right)}{\partial x} + \frac{\partial \left(u_x \tau_{yx}\right)}{\partial y} + \frac{\partial \left(u_x \tau_{zx}\right)}{\partial z}\right] \Delta x \Delta y \Delta z \end{equation}
    
    \begin{equation} \text{Y: } \left[\frac{\partial \left(u_y \tau_{xy}\right)}{\partial x} + \frac{\partial \left(u_y \tau_{yy}\right)}{\partial y} + \frac{\partial \left(u_y \tau_{zy}\right)}{\partial z}\right] \Delta x \Delta y \Delta z \end{equation}
    
    \begin{equation} \text{Z: } \left[\frac{\partial \left(u_z \tau_{xz}\right)}{\partial x} + \frac{\partial \left(u_z \tau_{yz}\right)}{\partial y} + \frac{\partial \left(u_z \tau_{zz}\right)}{\partial z}\right] \Delta x \Delta y \Delta z \end{equation}
    
* The rate of work done on the fluid element by the pressure force is also given the product of the relevant velocity and the force:
    
    \begin{equation} \text{X: } \left[-\frac{\partial \left(u_x p\right)}{\partial x}\right]\Delta x \Delta y \Delta z \end{equation}
    
    \begin{equation} \text{Y: } \left[-\frac{\partial \left(u_y p\right)}{\partial y}\right]\Delta x \Delta y \Delta z \end{equation}
    
    \begin{equation} \text{Z: } \left[-\frac{\partial \left(u_z p\right)}{\partial z}\right]\Delta x \Delta y \Delta z \end{equation}
    
* The rate of work done by other forces including gravity is given by:

    \begin{equation} \text{X: } \left[S_x u_x\right]\Delta x \Delta y \Delta z\end{equation}
    
    \begin{equation} \text{Y: } \left[S_y u_y\right]\Delta x \Delta y \Delta z\end{equation}
    
    \begin{equation} \text{Z: } \left[S_z u_z\right]\Delta x \Delta y \Delta z\end{equation}
    
* Combining all terms:
    
    \begin{equation} \rho \Delta x \Delta y \Delta z \frac{DE}{Dt} = -\left[\frac{\partial \left(u_x p\right)}{\partial x} + \frac{\partial \left(u_y p\right)}{\partial y} +  \frac{\partial \left(u_z p\right)}{\partial z}\right]\Delta x \Delta y \Delta z  - \left[ \frac{\partial}{\partial x} \left(k \frac{\partial T}{\partial x}\right) +  \frac{\partial}{\partial y} \left(k \frac{\partial T}{\partial y}\right) + \frac{\partial}{\partial z} \left(k \frac{\partial T}{\partial z}\right)\right]\Delta x \Delta y \Delta z + \left[\frac{\partial \left(u_x \tau_{xx}\right)}{\partial x} + \frac{\partial \left(u_x \tau_{yx}\right)}{\partial y} + \frac{\partial \left(u_x \tau_{zx}\right)}{\partial z} + \frac{\partial \left(u_y \tau_{xy}\right)}{\partial x} + \frac{\partial \left(u_y \tau_{yy}\right)}{\partial y} + \frac{\partial \left(u_y \tau_{zy}\right)}{\partial z} + \frac{\partial \left(u_z \tau_{xz}\right)}{\partial x} + \frac{\partial \left(u_z \tau_{yz}\right)}{\partial y} + \frac{\partial \left(u_z \tau_{zz}\right)}{\partial z} \right] \Delta x \Delta y \Delta z + \left[S_x u_x + S_y u_y + S_z u_z\right]\Delta x \Delta y \Delta z \end{equation}
    
* Dividing through by the volume of the fluid element:
    
    \begin{equation} \rho \frac{DE}{Dt} = -\left[\frac{\partial \left(u_x p\right)}{\partial x} + \frac{\partial \left(u_y p\right)}{\partial y} +  \frac{\partial \left(u_z p\right)}{\partial z}\right]  - \left[ \frac{\partial}{\partial x} \left(k \frac{\partial T}{\partial x}\right) +  \frac{\partial}{\partial y} \left(k \frac{\partial T}{\partial y}\right) + \frac{\partial}{\partial z} \left(k \frac{\partial T}{\partial z}\right)\right]+ \left[\frac{\partial \left(u_x \tau_{xx}\right)}{\partial x} + \frac{\partial \left(u_x \tau_{yx}\right)}{\partial y} + \frac{\partial \left(u_x \tau_{zx}\right)}{\partial z} + \frac{\partial \left(u_y \tau_{xy}\right)}{\partial x} + \frac{\partial \left(u_y \tau_{yy}\right)}{\partial y} + \frac{\partial \left(u_y \tau_{zy}\right)}{\partial z} + \frac{\partial \left(u_z \tau_{xz}\right)}{\partial x} + \frac{\partial \left(u_z \tau_{yz}\right)}{\partial y} + \frac{\partial \left(u_z \tau_{zz}\right)}{\partial z} \right] + \left[S_x u_x + S_y u_y + S_z u_z\right] \end{equation}
    
* The specific energy *E* usually comprises of internal energy *i*, kinetic energy $\frac{1}{2}\left(u_x^2 + u_y^2 + u_z^2\right)$, and gravitational potential energy. Gravitational effects have been lumped into other sources $S$. To obtain an equation for the internal energy, we need to remove kinetic energy contributions to the specific energy. This can be done by multiplying the momentum equations with their velocity components, summing them up and subtracting the result from the above equation. The sum of the product of momentum equations and their velocity components is given by:

    \begin{equation} \rho \frac{D\left[\frac{1}{2}\left(u_x^2 + u_y^2 + u_z^2\right)\right]}{Dt} = -\mathbf{u} \cdot \nabla p + u_x \left(\frac{\partial \tau_{xx}}{\partial x} + \frac{\partial \tau_{yx}}{\partial y} + \frac{\partial \tau_{zx}}{\partial z}\right) + u_y \left(\frac{\partial \tau_{xy}}{\partial x} + \frac{\partial \tau_{yy}}{\partial y} + \frac{\partial \tau_{zy}}{\partial z}\right)  + u_z \left(\frac{\partial \tau_{xz}}{\partial x} + \frac{\partial \tau_{yz}}{\partial y} + \frac{\partial \tau_{zz}}{\partial z}\right) + \mathbf{u} \cdot \mathbf{S}\end{equation}

* After subtracting the above equation from the specific energy equation, we get the following equation for internal energy:

    \begin{equation} \rho \frac{Di}{Dt} = -p\nabla \cdot \mathbf{u} + \nabla \cdot \left(k\nabla T\right) + \tau_{xx}\frac{\partial u_x}{\partial x} + \tau_{yx}\frac{\partial u_x}{\partial y} + \tau_{zx} \frac{\partial u_x}{\partial z} + \tau_{xy} \frac{\partial u_y}{\partial x} + \tau_{yy} \frac{\partial u_y}{\partial y} + \tau_{zy} \frac{\partial u_y}{\partial z} + \tau_{xz} \frac{\partial u_z}{\partial x} + \tau_{yz} \frac{\partial u_z}{\partial y} + \tau_{zz} \frac{\partial u_z}{\partial z} + \mathbf{S}_i \end{equation}
    
## GOVERNING EQUATIONS FOR NEWTONIAN FLUIDS
***

### Continuity Equation


* The continuity equation remains unchanged as shown above:

    \begin{equation}\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z} = 0\end{equation}
    
### Momentum Balance Equations


* For Newtonian fluids, we can use Stokes' relations for viscous stresses:

    - Linear transformations: 
        
        \begin{equation}\tau_{xx} = 2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u}\end{equation}
        
        \begin{equation}\tau_{yy} = 2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u}\end{equation}
        
        \begin{equation}\tau_{zz} = 2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u}\end{equation}

    - Volumetric transformations: 
        
        \begin{equation}\tau_{xy} = \tau_{yx} = \mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\end{equation}
        
        \begin{equation}\tau_{xz} = \tau_{zx} = \mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\end{equation}
        
        \begin{equation}\tau_{yz} = \tau_{zy} = \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\end{equation}

* Plugging Stokes' relations into the general momentum balance equations:
    
    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial}{\partial x}\left[2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right]  + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \rho g_x + S_x\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \rho g_y + S_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \frac{\partial}{\partial z}\left[2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u} \right] + \rho g_z + S_z\end{equation}

* Re-arranging:

    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z\end{equation}

* In more compact form:

    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z\end{equation}
    
### Energy Equation

* Substituting the Stokes relations for the viscous stresses into the general energy equation:

    \begin{equation} \rho \frac{Di}{Dt} = -p\nabla \cdot \mathbf{u} + \nabla \cdot \left(k\nabla T\right) + \left(2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u}\right)\frac{\partial u_x}{\partial x} + \mu\left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right) \frac{\partial u_x}{\partial y} + \mu \left(\frac{\partial u_z}{\partial x} + \frac{\partial u_x}{\partial z}\right) \frac{\partial u_x}{\partial z} + \mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right) \frac{\partial u_y}{\partial x} + \left(2\mu\frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u}\right) \frac{\partial u_y}{\partial y} + \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\frac{\partial u_y}{\partial z} + \mu\left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\frac{\partial u_z}{\partial x} + \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\frac{\partial u_z}{\partial y} + \left(2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u}\right) \frac{\partial u_z}{\partial z} + \mathbf{S}_i\end{equation}

## GOVERNING EQUATIONS FOR INCOMPRESSIBLE FLOWS
***

### Continuity Equation


* For incompressible flows, the density is constant, so we can set $\frac{\partial \rho}{\partial t} = 0$:

    \begin{equation}\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z} = 0\end{equation}

* In more compact form:

    \begin{equation}\nabla \cdot \mathbf{u} = 0\end{equation}
    
### Momentum Balance Equations


* We can use the continuity equation for incompressible flows shown above to simplify the momentum balance equations for Newtonian fluids by setting $\nabla \cdot \mathbf{u} = 0$. This should result in the following:

    \begin{equation}\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \rho g_x + S_x\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \rho g_y + S_y\end{equation}

    \begin{equation}\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \rho g_z + S_z\end{equation}
    
### Energy Equation

* For incompressible flows, $i = c T$ and $\nabla \cdot \mathbf{u} = 0$. 

    \begin{equation} \rho c \frac{DT}{Dt} =\nabla \cdot \left(k\nabla T\right) + \left(2\mu \frac{\partial u_x}{\partial x}\right)\frac{\partial u_x}{\partial x} + \mu\left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right) \frac{\partial u_x}{\partial y} + \mu \left(\frac{\partial u_z}{\partial x} + \frac{\partial u_x}{\partial z}\right) \frac{\partial u_x}{\partial z} + \mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right) \frac{\partial u_y}{\partial x} + \left(2\mu\frac{\partial u_y}{\partial y}\right) \frac{\partial u_y}{\partial y} + \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\frac{\partial u_y}{\partial z} + \mu\left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\frac{\partial u_z}{\partial x} + \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\frac{\partial u_z}{\partial y} + \left(2\mu \frac{\partial u_z}{\partial z}\right) \frac{\partial u_z}{\partial z} + \mathbf{S}_i\end{equation}
