# DISCRETIZATION OF TRANSIENT TERM
***

* The Governing equation for a transient 1-D diffusion process is:

    \begin{equation} \frac{\partial \left(\rho \phi\right)}{\partial t} -  \frac{\partial}{\partial x} \left(\Gamma \frac{\partial \phi}{\partial x}\right) = 0 \label{eqn1} \tag{1}\end{equation}
    
* Integrating equation \ref{eqn1} with respect to time and over the control volume:

    \begin{equation} \int_t^{t+\Delta t} \frac{\partial}{\partial t} \left(\int_{cv}\rho \phi dV\right) dt -  \int_t^{t+\Delta t} \int_{cv} \frac{\partial}{\partial x} \left(\Gamma \frac{\partial \phi}{\partial x}\right) dV dt = 0 \label{eqn2} \tag{2}\end{equation}

    \begin{equation} \int_t^{t+\Delta t} \frac{\partial}{\partial t} \left(\rho \phi \Delta x \right) dt -  \int_t^{t+\Delta t} \int_{cv} \frac{\partial}{\partial x} \left(\Gamma \frac{\partial \phi}{\partial x}\right) dV dt = 0 \label{eqn3} \tag{3}\end{equation}
    
* Applying Gauss divergence theorem to the diffusion term:

    \begin{equation} \int_t^{t+\Delta t} \frac{\partial}{\partial t} \left(\rho \phi \Delta x \right) dt -  \int_t^{t+\Delta t} \int_{cs} \mathbf{n} \cdot \left(\Gamma \frac{\partial \phi}{\partial x}\right) dS dt = 0 \label{eqn4} \tag{4}\end{equation}
    
    \begin{equation} \int_t^{t+\Delta t} \frac{\partial}{\partial t} \left[\rho \phi \Delta x \right] dt +  \int_t^{t+\Delta t} \left[\Gamma_w \left(\frac{\phi_P - \phi_W}{\delta x_w}\right) - \Gamma_e \left(\frac{\phi_E - \phi_P}{\delta x_e}\right)\right] dt = 0 \label{eqn5} \tag{5}\end{equation}

* To estimate the time integral of the diffusion term in equation \ref{eqn5}, we need to decide whether the integral should be based on values from the previous or current time step knowing that the values at current time step have not yet been calculated. To make this general, we will use weighting factors to blend previous and current time values. We shall also require the weighting factors to add up to one. We will use the superscript 0 to denote a value at a previous time step. Any dependent variable $\phi$ with no superscript belongs to the current time step.

    \begin{equation} \frac{\rho \Delta x}{\Delta t} \left(\phi_P - \phi_P^0\right) + f\left[\Gamma_w \left(\frac{\phi_P - \phi_W}{\delta x_w}\right) - \Gamma_e \left(\frac{\phi_E - \phi_P}{\delta x_e}\right)\right]  + \left(1 - f\right)\left[\Gamma_w \left(\frac{\phi_P^0 - \phi_W^0}{\delta x_w}\right) - \Gamma_e \left(\frac{\phi_E^0 - \phi_P^0}{\delta x_e}\right)\right] = 0  \label{eqn6} \tag{6}\end{equation}
    
* Re-arranging:
    
    \begin{equation} \left[\frac{\rho \Delta x}{\Delta t} + f\left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right)\right]\phi_P = f\left(\frac{\Gamma_w}{\delta x_w}\right)\phi_W + f\left(\frac{\Gamma_e}{\delta x_e}\right)\phi_E +  \left[\frac{\rho \Delta x}{\Delta t} - \left(1-f\right)\frac{\Gamma_w}{\delta x_w} - \left(1 - f\right)\frac{\Gamma_e}{\delta x_e}\right]\phi_P^0 + \left(1 - f\right)\left[\frac{\Gamma_w}{\delta x_w} \phi_W^0 +  \frac{\Gamma_e}{\delta x_e} \phi_E^0\right] = 0 \label{eqn7} \tag{7}\end{equation}
    
## FULLY IMPLICIT SCHEME
***

* An implicit scheme uses values at current time step to compute the diffusion term. This implies the weighting factor $f = 1$. Applying this to equation \ref{eqn7}:

    \begin{equation} a_p \phi_P = a_w \phi_W + a_e \phi_E + a_p^0 \phi_P^0 \label{eqn8} \tag{8} \end{equation}
    
    \begin{equation} a_p = \frac{\rho \Delta x}{\Delta t} + \frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e} \label{eqn9} \tag{9}\end{equation}
    
    \begin{equation} a_w = \frac{\Gamma_w}{\delta x_w} \label{eqn10} \tag{10}\end{equation}
    
    \begin{equation} a_e = \frac{\Gamma_e}{\delta x_e} \label{eqn11} \tag{11}\end{equation}
    
    \begin{equation} a_p^0 = \frac{\rho \Delta x}{\Delta t} \label{eqn12} \tag{12}\end{equation}

## FULLY EXPLICIT SCHEME
***

* For an explicit scheme, the values at previous time step are used to compute the diffusion term. This means the weighting factor $f = 0$ and applying this to equation \ref{eqn7} should yield:

    \begin{equation} a_p \phi_P = a_w \phi_W^0 + a_e \phi_E^0 + a_p^0 \phi_P^0 \label{eqn13} \tag{13} \end{equation}
    
    \begin{equation} a_p = \frac{\rho \Delta x}{\Delta t} \label{eqn14} \tag{14} \end{equation}
    
    \begin{equation} a_w = \frac{\Gamma_w}{\delta x_w} \label{eqn15} \tag{15}\end{equation}
    
    \begin{equation} a_e = \frac{\Gamma_e}{\delta x_e} \label{eqn16} \tag{16}\end{equation}
    
    \begin{equation} a_p^0 = \frac{\rho \Delta x}{\Delta t} - \left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right) \label{eqn17} \tag{17}\end{equation}

* For stability, coefficients in equation \ref{eqn13} need to be all positive. However, it is possible for $a_p^0$ to be negative if its first term $\frac{\rho \Delta x}{\Delta t}$ is smaller the second term $\left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right)$. To enforce stability, the time step $\Delta t$ needs to be selected to ensure that $a_p^0$ is always positive. This stability criterion is represented below:
    
    \begin{equation} \frac{\rho \Delta x}{\Delta t} > \left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right) \label{eqn18} \tag{18} \end{equation}
    
* For constant diffusion coefficient and uniform grid:
    
    \begin{equation} \frac{\rho \Delta x}{\Delta t} > \frac{2\Gamma}{\Delta x} \label{eqn19} \tag{19}\end{equation}
    
    \begin{equation} \Delta t < \frac{\rho \Delta x^2}{2\Gamma} \label{eqn20} \tag{20} \end{equation}
    
## CRANK-NICHOLSON
***

* Any scheme whose weighting factor is not zero is by definition an implicit scheme. The Crank-Nicholson scheme whose weighting factor is 0.5 is thus an implicit scheme. If we apply the f = 0.5 to equation \ref{eqn7}, we should get:

    \begin{equation} a_p \phi_P = a_w \phi_W + a_e \phi_E + a_p^0\phi_P^0 + a_w\phi_W^0 + a_e \phi_E^0 \label{eqn21} \tag{21}\end{equation}
    
    \begin{equation} a_p = \frac{\rho \Delta x}{\Delta t} + 0.5\left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right) \label{eqn22} \tag{22}\end{equation}
    
    \begin{equation} a_w = 0.5\frac{\Gamma_w}{\delta x_w} \label{eqn23} \tag{23}\end{equation}
    
    \begin{equation} a_e = 0.5\frac{\Gamma_e}{\delta x_e} \label{eqn24} \tag{24}\end{equation}
    
    \begin{equation} a_p^0 = \frac{\rho \Delta x}{\Delta t} - 0.5\left(\frac{\Gamma_w}{\delta x_w} + \frac{\Gamma_e}{\delta x_e}\right) \label{eqn25} \tag{25}\end{equation}
    
* The Crank-Nicholson scheme is unconditionally stable despite the fact that $a_p^0$ can be negative. However, it can produce oscillatory solutions but these oscillations eventually die out due to the unconditional stability. The posibility of negative $a_p^0$ means that physically unrealistic solutions can be obtained.
